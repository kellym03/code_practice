```python
import codecademylib3
import pandas as pd

ad_clicks = pd.read_csv('ad_clicks.csv')
print(ad_clicks.head(10))

#How many views (i.e., rows of the table) came from each utm_source?
platform_views = ad_clicks.groupby('utm_source').user_id.count().reset_index()
print(platform_views)

#Create a new column called is_click, which is True if ad_click_timestamp is not null and False otherwise.
ad_clicks['is_click'] = ad_clicks.apply(
    lambda row: True if pd.notnull(row.ad_click_timestamp) else False, 
    axis=1
) #More efficient is # ad_clicks['clicked_ad'] = ad_clicks.ad_click_timestamp.notnull()
print(ad_clicks)

#The percent of people who clicked on ads from each utm_source.
#Start by grouping by utm_source and is_click and counting the number of user_id‘s in each of those groups.
clicks_by_source = ad_clicks.groupby(['utm_source','is_click']).user_id.count().reset_index()
print(clicks_by_source)

#Pivot to make it easier to calculate the percent of people who clicked on ads from each utm_source.
clicks_pivot = clicks_by_source.pivot(
  columns = 'is_click',
  index= 'utm_source',
  values= 'user_id'
).reset_index()

print(clicks_pivot)

#Calculate the percent of people who clicked on ads from each utm_source and add it as a new column to clicks_pivot.
clicks_pivot['percent_clicked'] = clicks_pivot[True] / (clicks_pivot[False] + clicks_pivot[True])
print(clicks_pivot)

ab_group_count = ad_clicks.groupby('experimental_group').user_id.count().reset_index()
print(ab_group_count)

ab_click_split = ad_clicks.groupby(['experimental_group','is_click']).user_id.count().reset_index()
ab_click_split_pivot = ab_click_split.pivot(
  columns = 'is_click',
  index = 'experimental_group',
  values = 'user_id',
)

ab_click_split_pivot['percent_clicked'] = ab_click_split_pivot[True] / (ab_click_split_pivot[False] + ab_click_split_pivot[True])
print(ab_click_split_pivot)

a_clicks = ad_clicks[ad_clicks.experimental_group == 'A']
#print(a_clicks)
b_clicks = ad_clicks[ad_clicks.experimental_group == 'B']

a_clicks_day = a_clicks.groupby(['day','is_click']).user_id.count().reset_index()
a_clicks_day_pivot = a_clicks_day.pivot(
  columns = 'is_click',
  index = 'day',
  values = 'user_id',
).reset_index()
a_clicks_day_pivot['percent_clicked'] = a_clicks_day_pivot[True] / (a_clicks_day_pivot[False] + a_clicks_day_pivot[True])
print(a_clicks_day_pivot)

b_clicks_day = b_clicks.groupby(['day','is_click']).user_id.count().reset_index()
b_clicks_day_pivot = b_clicks_day.pivot(
  columns = 'is_click',
  index = 'day',
  values = 'user_id',
).reset_index()
b_clicks_day_pivot['percent_clicked'] = b_clicks_day_pivot[True] / (b_clicks_day_pivot[False] + b_clicks_day_pivot[True])
print(b_clicks_day_pivot)

#Create a new DataFrame that contains the percent of people who clicked on ads from group A and group B by day of the week.
combined_table_percent_clicks = pd.DataFrame({'Day of Week':a_clicks_day_pivot['day'], 'Group_A_clicks': a_clicks_day_pivot['percent_clicked'], 'Group_B_clicks': b_clicks_day_pivot['percent_clicked']}) #

print(combined_table_percent_clicks)
#Based on the combined table, it looks like group A had a higher percentage of clicks than group B on nearly every day of the week. This suggests that the ads shown to group A may have been more effective at encouraging users to click on them compared to the ads shown to group B. However, it's important to note that this is just an observational analysis and further statistical testing would be needed to determine if the difference in click-through rates between the two groups is statistically significant.
```
